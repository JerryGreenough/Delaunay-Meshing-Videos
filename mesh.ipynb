<h2>Create a Delaunay Triangulation</h2>
<p>
A counter-clockwise oriented set of boundary nodes is defined. The addBoundaryLoop() method in the code
that follows takes this list as an input and furnishes the mesh object m1 with a mesh. The set of boundary nodes is
defined in counter-clockwise order with boundary edges defined between consecutive nodes of the list. 
A final boundary edge is inserted between the final node of the list and the start node. The edges that are defined in this
way are depicted below in black, with the node numbers displayed in blue.
</p>

```python
from draw import plot_nodelist

# Create the counter-clockwise loop of boundary nodes.

nodeList = [[0,0], 
            [1,0], [2,0], [2,0.4], [3,0.4], [3,0], [5,0], [5,1], \
            [3,1], [3, 0.6], [2, 0.6], [2,1], [1,1], \
            [1,2], [2,2], [2,2.4], [3,2.4], [3,2], [5,2], [5,3], \
            [3,3], [3, 2.6], [2, 2.6], [2,3], [1,3], \
            [1,4], [2,4], [2,4.4], [3,4.4], [3,4], [5,4], [5,5], \
            [3,5], [3, 4.6], [2, 4.6], [2,5], [0,5]]

plot_nodelist(nodeList)
```

```python
# Create the counter-clockwise loop of boundary nodes.

nodeList = [[0,0], 
            [1,0], [2,0], [2,0.4], [3,0.4], [3,0], [5,0], [5,1], \
            [3,1], [3, 0.6], [2, 0.6], [2,1], [1,1], \
            [1,2], [2,2], [2,2.4], [3,2.4], [3,2], [5,2], [5,3], \
            [3,3], [3, 2.6], [2, 2.6], [2,3], [1,3], \
            [1,4], [2,4], [2,4.4], [3,4.4], [3,4], [5,4], [5,5], \
            [3,5], [3, 4.6], [2, 4.6], [2,5], [0,5]]

# Create a mesh object.

m1 = mesh()

# Add the boundary loop data and mesh the interior.

m1.addBoundaryLoop(nodeList)     

# Plot the mesh. The boundary edges are displayed in yellow.
            
m1.plot(labels=False, arrows=False)
```

<h2>Create an Animation of a Delaunay Triangulation</h2>
<p>
The addBoundaryLoop() method that was used in the preceding code is replaced by the addBoundaryLoopWithVideo() in the code
that follows. This ensures that the mesh resulting from each iteration of the Delaunay meshing algorithm is recorded in
the mp4 file 'm1.mp4'.
</p>

```python
nodeList = [[0,0], 
            [1,0], [2,0], [2,0.4], [3,0.4], [3,0], [5,0], [5,1], \
            [3,1], [3, 0.6], [2, 0.6], [2,1], [1,1], \
            [1,2], [2,2], [2,2.4], [3,2.4], [3,2], [5,2], [5,3], \
            [3,3], [3, 2.6], [2, 2.6], [2,3], [1,3], \
            [1,4], [2,4], [2,4.4], [3,4.4], [3,4], [5,4], [5,5], \
            [3,5], [3, 4.6], [2, 4.6], [2,5], [0,5]]
            
m1 = mesh()

# Specify a path for the mp4 file.

fpath = ".\m1.mp4"

# Initialize the video plotting parameters.

m1.vplot_init(fpath, figsize=(6,6))

# Mesh the interior of the boundary with triangles, recording (as a frame
# in the video file m1.mp4) the state
# of the mesh after each boundary edge has been added.  

m1.addBoundaryLoopWithVideo(nodeList)
```

```python
import numpy as np

aa = np.array([[1,0], [1,1]])

bb = aa.copy()
for i in range(0,3):
    bb = bb + np.array([1,0])
    aa = np.concatenate([aa, bb])
    
bb = aa + np.array([0,1.3])
end0 = np.array([[0, 2.3], [0,0]])
end1 = np.array([[5,0], [5,2.3]])

nodeList = np.concatenate([aa, end1, bb[-1::-1], end0])

m4 = mesh()

m4.addBoundaryLoop(nodeList)

m4.plot(figsize=(12,12))

m4 = mesh()

m4.addNodes(nodeList)

fpath = ".\m4.mp4"
m4.vplot_init(fpath, figsize=(6,6))
m4.addBoundaryLoopWithVideo(nodeList)
```
