```python
from matplotlib.animation import FFMpegWriter
import matplotlib.pyplot as plt

writer = FFMpegWriter(fps=2)

figsize=(6,6)
fig, axs = plt.subplots(1,1, figsize=figsize)

m6 = mesh()

nodeList = [[0.0, 0.0], [2.0, 0.0], [1.0, 3.0], [0.5, 2.5], [-0.3, 1.9], [-0.3, 0.4]]
m6.addNodes(nodeList)

from pathlib import Path

fle = Path('.\writer_test.mp4')
fle.touch(exist_ok=True)

with writer.saving(fig, "C:\\Users\\jgree\\Documents\\Projects\\Geometry\\delaunay_animation\\writer_test.mp4", 200):

    for ixn, nn in enumerate(nodeList[:-1]):
        m6.insertBoundaryEdge([ixn, ixn+1])
        m6.vplot(writer, fig, axs, labels=False)
    
    m6.insertBoundaryEdge([len(nodeList)-1, 0])
    m6.vplot(writer, fig, axs, labels=False)
            
    m6.removeNonFeatureBoundaryEdges()
    m6.vplot(writer, fig, axs, labels=False)
```
